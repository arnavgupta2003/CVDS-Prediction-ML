```python
import numpy as np
import pandas as pd

data = pd.read_csv("cardio_train.csv")
```

## Plots


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Corelation HeatMap
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Scatter plot //Computationally Intensive, Took 7m 30s
# Create pair plots for each column
sns.pairplot(data, diag_kind='kde', hue='cardio', palette='coolwarm', markers=['o', 's'])
plt.suptitle("Pair Plots of Each Column", y=1.02)
plt.show()
plt.savefig("PairPlot.png")
```

```python
sns.set(style="whitegrid")

# Plot 1: Countplot of 'gender'
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=data, hue='cardio', palette='coolwarm')
plt.title("Countplot of Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Plot 2: Distribution of 'age' with histograms
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True, color='blue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot 3: Pairplot of selected numerical columns // Without ap_hi & ap_low
subset_columns = ["age", "gender","height", "weight", "cholesterol", "gluc", "smoke","active", "PP","MAP","cardio"]
sns.pairplot(data[subset_columns], hue='cardio', palette='coolwarm', markers=['o', 's'], diag_kind='kde')
plt.suptitle("Pair Plots of Selected Columns", y=1.02)
plt.show()

```

```python
# Plot 4: Correlation heatmap // Without ap_hi & ap_low
correlation_matrix = data[subset_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
